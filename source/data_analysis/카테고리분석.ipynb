```python
root = Path("../../data/books/")
fn_book = root / Path("삼정일계표('23-05월).xlsx")
fn_total = root / Path("22년 현금+카드정산(최종230313).xlsx")
df_book = pd.read_excel(fn_book, sheet_name=['입금계정 일람표', '출금계정 일람표'])
df_tran = pd.read_excel(fn_total, skiprows=1)
```
